```
a = 3
b = 0
def add(a,b):
  print(f"{a}+{b} = {a+b}")

def sub(a,b):
  print(f"{a}-{b} = {a - b}")

def mul(a,b):
  print(f"{a}*{b} = {a * b}")

def div(a,b):
  print(f"{a}/{b} = {a / b}")

operation = [add,sub,mul,div]

for op in operation:
  try:
    op(a,b)
  except Exception as e:
    print(e.args)
```
